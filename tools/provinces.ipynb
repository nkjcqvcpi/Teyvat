```python
%matplotlib inline

import cv2 as cv
import numpy as np
import random as rng
from matplotlib import pyplot as plt

ori_map = cv.imread("map2201(useful).png")
h, w, _ = ori_map.shape
canny_map = cv.Canny(ori_map, 50, 150)
plt.imshow(cv.cvtColor(canny_map, cv.COLOR_BGR2RGB))
```

```python
contours, hierarchy = cv.findContours(canny_map, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
drawing = np.zeros((canny_map.shape[0], canny_map.shape[1], 3), dtype=np.uint8)
for i in range(len(contours)):
    color = (rng.randint(0,256), rng.randint(0,256), rng.randint(0,256))
    cv.drawContours(drawing, contours, i, color, 2, cv.LINE_8, hierarchy, 0)
plt.imshow(drawing)
```

```python
drawing = cv.polylines(drawing, contours, True, (255, 255, 255))
drawing = cv.fillPoly(drawing, contours, (255, 255, 255))
plt.imshow(drawing)
```
